```python
import pandas as pd
import numpy as np
```

```python
df_courses = pd.read_csv('courses.csv')
df_courses
```

```python
df_marks = pd.read_csv('marks.csv')
df_marks
```

```python
df_students = pd.read_csv('students.csv')
df_students
```

```python
df_students.rename(columns ={'id':'student_id'}, inplace =True )
df_students
```

```python
df_std_marks = pd.merge(df_marks,df_students, on="student_id", how="inner")
df_std_marks.rename(columns ={'name':'student_name'}, inplace =True )
df_std_marks
```

```python
df_tests = pd.read_csv('tests.csv')
df_tests
```

```python
df_courses.rename(columns ={'id':'course_id'}, inplace =True )
df_courses
```

```python
df_crs_tests = pd.merge(df_courses,df_tests, on="course_id", how="inner")
df_crs_tests
```

```python
df_crs_tests.rename(columns ={'id':'test_id'}, inplace =True )
df_crs_tests.rename(columns ={'name':'course_name'}, inplace =True )
df_crs_tests
```

```python
df_all = pd.merge(df_std_marks,df_crs_tests, on="test_id", how="inner")
df_all
```

```python
column_names = ['student_id', 'student_name', 'mark', 'course_id',
       'course_name', 'teacher', 'weight']
df_ALL = pd.DataFrame(df_all, columns=column_names)
df_ALL
```

```python
def averageOfList(num):
    sumOfNumbers = 0
    for t in num:
        sumOfNumbers = sumOfNumbers + t

    avg = sumOfNumbers / len(num)
    return avg

print("The average mark of student A for Biology course is", round(averageOfList([78, 87, 95]),1), "%")
print("The average mark of student A for History course is", round(averageOfList([32, 65]),1), "%")
print("The average mark of student A for Math course is", round(averageOfList([78, 40]),1), "%")
#print("The total average mark of student A for all courses is", round(averageOfList([78, 87, 95, 32, 65, 78, 40]),1), "%")
```
